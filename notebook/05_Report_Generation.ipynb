```python
import os
import sys
os.chdir(r"D:\tech\ADVT_python\vehicle_damage_detection")
PROJECT_PATH = os.getcwd()
if PROJECT_PATH not in sys.path:
    sys.path.append(PROJECT_PATH)
print(PROJECT_PATH)
```

```python
# imports
from src.detector import DamageDetector
from src.assessment import DamageAssessment
from src.report import ReportGenerator
import pandas as pd
```

```python
# load models
MODEL_PATH = "./models/best.pt"
detector = DamageDetector(MODEL_PATH)
assessment_engine = DamageAssessment()
report_generator = ReportGenerator()
print("Model Loaded Successfully.")
```

```python
IMAGE_PATH = "./data/test/images/000040.jpg"
print(IMAGE_PATH)
```

```python
# detection
annotated_image, detections = detector.detect(IMAGE_PATH)
print(f"Detected Damages : {len(detections)}")
```

```python
#assessment
assessment = assessment_engine.assess(detections)
print("Assessment Completed.")
```

```python
PREDICTION_IMAGE = "./reports/figures/prediction.jpg"
detector.save_prediction(
    annotated_image,
    PREDICTION_IMAGE
)
print("Prediction image saved.")
```

```python
#  vehicle summary
summary = assessment["summary"]
print("VEHICLE SUMMARY")
for key, value in summary.items():
    print(f"{key.replace('_',' ').title():25}: {value}")
```

```python
# damage table
damage_df = report_generator.to_dataframe(assessment)
damage_df["Confidence (%)"] = (
    damage_df["Confidence (%)"]
    .astype(float)
    .map(lambda x: f"{x:.1f}%")
)
display(damage_df)
```

```python
csv_path = report_generator.export_csv(
    assessment,
    filename="damage_report.csv"
)
print("CSV Saved Successfully")
print(csv_path)
```

```python
pdf_path = report_generator.export_pdf(
    assessment,
    filename="damage_report.pdf",
    prediction_image_path=PREDICTION_IMAGE
)
print("PDF Saved Successfully")
print(pdf_path)
```

```python
#verifying saved files
from pathlib import Path

print("\nGenerated Files\n")

for folder in ["reports/csv", "reports/pdf", "reports/figures"]:
    print(f"\n{folder}")

    for file in sorted(Path(folder).glob("*")):
        print("   ", file.name)
```
